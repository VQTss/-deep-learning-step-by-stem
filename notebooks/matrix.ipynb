```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
from IPython.display import Image as IPImage, display
```

```python
import sys, os
sys.path.insert(0, os.path.abspath(os.path.join(os.getcwd(), "..")))
from src.utils.matrix import matrix_transpose
```

```python
# Load image with OpenCV (returns BGR ndarray)
image_path = "../assets/dog.jpg"
img_bgr = cv2.imread(image_path)
# Convert BGR -> RGB for matplotlib display
img_array = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
```

```python
print("Shape:", img_array.shape)
```

```python
plt.figure(figsize=(5, 6))
plt.imshow(img_array)
plt.title('Original', fontsize=14, fontweight='bold')
plt.axis('off')
plt.show()
```

```python
# 1) Algebraic transpose via src.utils.matrix.matrix_transpose
# matrix_transpose returns A.T; for an (H, W, 3) image that gives (3, W, H)
img_transposed_array = matrix_transpose(img_array)

print('Shape before transpose :', img_array.shape)
print('Shape apply transpose :', img_transposed_array.shape)

# Re-arrange axes from (channels, W, H) back to (W, H, channels) for matplotlib
img_transposed_display = np.transpose(img_transposed_array, (1, 2, 0))

fig, axes = plt.subplots(1, 2, figsize=(12, 5))
axes[0].imshow(img_array)
axes[0].set_title(f'Original\n{img_array.shape[1]} x {img_array.shape[0]} pixels', fontsize=13, fontweight='bold', color='#27ae60')
axes[0].axis('off')

axes[1].imshow(img_transposed_display)
axes[1].set_title(f'Apply Transpose (matrix_transpose)\n{img_transposed_display.shape[1]} x {img_transposed_display.shape[0]} pixels', fontsize=13, fontweight='bold', color='#c0392b')
axes[1].axis('off')

plt.suptitle('Transpose', fontsize=15, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()
```

```python
# 2) True rotation with OpenCV (what you usually want visually)
img_rotated = cv2.rotate(img_array, cv2.ROTATE_90_CLOCKWISE)
print('Shape after rotate:', img_rotated.shape)

fig, axes = plt.subplots(1, 2, figsize=(12, 5))
axes[0].imshow(img_array)
axes[0].set_title('Original', fontsize=13, fontweight='bold')
axes[0].axis('off')

axes[1].imshow(img_rotated)
axes[1].set_title('Rotated 90\u00b0 (clockwise)', fontsize=13, fontweight='bold', color='#2980b9')
axes[1].axis('off')

plt.tight_layout()
plt.show()
```
